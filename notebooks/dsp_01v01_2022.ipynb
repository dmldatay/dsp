# Part 1: One Dimensional Signals

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(-2*np.pi, 2*np.pi, np.pi/10)
cos_wav = np.cos(t)
```

```python
plt.plot(t, cos_wav)
plt.title('Cosine Wave (Continuous)')
plt.xlabel('Period') 
plt.ylabel('Amplitude')
plt.grid()
```

```python
plt.stem(t, cos_wav)
plt.title('Coine Wave (Stem Plot)')
plt.xlabel('Period') 
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

```python
import os
import sys
sys.path.insert(0, 'ThinkDSP/code/') 
import thinkdsp
import matplotlib.pyplot as pyplot
import IPython
```

```python
# This file shows how to generate a sine wave using the ThinkDSP library.
import sys
sys.path.insert(0, 'ThinkDSP/code/') 
import thinkdsp
import IPython

# Generate wave
sin = thinkdsp.SinSignal(freq=400, amp=0.5) # Where freq is the frequency and amp is the amplitude of the sine wave
wave = sin.make_wave(duration=2, start=0, framerate=44100) # Turn the sin wave into an audio wave

# Plot wave
period = sin.period
segment = wave.segment(start=0, duration=period*3)
segment.plot()
pyplot.show()

# Play wave
wave.play()
IPython.display.Audio('sound.wav') # This is required on Google Colab due to compatibility issues
```

```python
wave = thinkdsp.read_wave('test.wav')

wave.plot()
pyplot.show()

# Play audio file
wave.play()
IPython.display.Audio('sound.wav')
```
